- Slicing : [*start*:*stop*:*step*]

- Fancy Indexing : [*list of index*]
- index, start, end의 값이 음수인 경우 배열의 뒤에서부터 indexing


### Slicing : [*start*:*stop*:*step*]

```python
# array1에 0부터 11까지의 1차원 배열 생성
array1 = np.arange(10)
print(array1)

# 앞에서 3개 요소 반환
print(array1[0:3])

# 앞에서 3번째부터 마지막 요소까지 반환
#print(array1[2::2])
print(array1[2:])

# 전체 반환
print(array1[:])

# 뒤에서 3번째 부터 반환
print(array1[-3:])

# 앞에서부터 뒤에서 3번째 요소까지 반환
print(array1[:-2:])

# 뒤에서부터 모두 반환
print(array1[::-1])
```

### Boolean Indexing : [*Boolean Condition*]

```python
array1d = np.arange(start=1, stop=10)
print(array1d)
# 5보다 큰 값만 출력
array2 = array1d[5<array1d]
print(array2)

# 5보다 크고 8보다 작은 값만 출력
array3 = array1d[(5 < array1d) & (array1d < 8)]
print(array3)
```

```python
boolean_indexes = np.array([False, False, False, False, False,  True,  True,  True,  True])
array3 = array1d[boolean_indexes]
print('불린 인덱스로 필터링 결과 :', array3)
```

### Fancy Indexing : [*list of index*]

```python
print(array1d)
# index 5, 1, 3, 2 순서대로 데이터 출력
array4 = array1d[[5, 1, 3, 2]]
print(array4)
```

- ndarray.sort() : 원 행렬 변경 → 반환값 없음
- numpy.argsort(*ndarray*) : 정렬 시의 index 순서 반환

```python
org_array = np.array([ 3, 1, 9, 5])

# np.sort( )로 정렬 후 원 행렬과 반환값 출력
sorted_array1 = np.sort(org_array)
print(org_array, sorted_array1)

# ndarray.sort( )로 정렬 후 원 행렬과 반환값 출력
# sorted_array2 = org_array.sort()
print(org_array, sorted_array2)

# np.argsort()로 인덱스만 반환
print(np.argsort(org_array))
```
